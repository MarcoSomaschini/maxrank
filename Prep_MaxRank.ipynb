```python
import pandas as pd

"""
Load the data
"""

datafile = 'data.csv'           # Path to your datafile
queryfile = 'query.csv'         # Path to your queryfile

df = pd.read_csv(datafile, index_col=0)
q = pd.read_csv(queryfile, index_col=0).index

df.head()
```

```python
"""
Invert the data values if the target optimal point is (0, ..., 0) instead of (1, ..., 1)
"""

df_cls = 1 - df
```

```python
"""
Create datafile
"""

n = df.shape[0]
dim = df.shape[1]

with open('data.txt', 'w+') as f:
    for r in range(n):
        f.write("{} ".format(r + 1))

        for c in range(dim):
            f.write("{:.6f} ".format(df.iloc[r][df.columns[c]] - 0.000001))
        for c in range(dim):
            f.write("{:.6f} ".format(df.iloc[r][df.columns[c]] + 0.000001))

        f.write("\n")
```

```python
"""
Create queryfile
"""

with open('query.txt', 'w+') as f:
    for idx in q:
        f.write("{}\n".format(df.index.get_loc(idx) + 1))
```
